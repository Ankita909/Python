**MATPLOTLIB**

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.plot([1,2,3],[4,5,7])
plt.show()
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.plot([1,2,8],[4,5,0])
plt.title("Ankita's Data")
plt.ylabel("This is Y-axis")
plt.xlabel("This is X-axis")
plt.show()
```

```python
# Plotting two lines in a single chart

from matplotlib import pyplot as plt
x=[10,3,40]
y=[77,8,98]
x2=[22,33,44]
y2=[33,55,66]
plt.plot(x,y)
plt.plot(x2,y2)
```

```python
# Change the style of the lines

from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot') # more styles are available on goggle, just search for matplotlib styles
x=[10,3,40]
y=[77,8,98]
x2=[22,33,44]
y2=[33,55,66]
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

```python
# Adding labels in the chart

from matplotlib import pyplot as plt

x=[10,3,40]
y=[77,8,98]
x2=[22,33,44]
y2=[33,55,66]
plt.plot(x,y,label="first plot")
plt.plot(x2,y2,label="second plot")
plt.legend()  # it is important to call this method legend() to get the labels displayed
plt.show()
```

```python
# Adding labels in the chart
# Changing the thickness of line

from matplotlib import pyplot as plt

x=[10,3,40]
y=[77,8,98]
x2=[22,33,44]
y2=[33,55,66]
plt.plot(x,y,label="first plot",linewidth=4)
plt.plot(x2,y2,label="second plot",linewidth=6)
plt.legend()  # it is important to call this method legend() to get the labels displayed
plt.show()
```

```python
# Now, we don't want to plot lines ,but want it scattered 

from matplotlib import pyplot as plt

x=[10,3,40]
y=[77,8,98]
x2=[22,33,44]
y2=[33,55,66]
plt.scatter(x,y,label="first plot",linewidth=4)
plt.scatter(x2,y2,label="second plot",linewidth=6)
plt.legend()  # it is important to call this method legend() to get the labels displayed
plt.show()
```

```python
# Histogram

from matplotlib import pyplot as plt

x=[10,3,40]
y=[77,8,98]
x2=[22,33,44]
y2=[33,55,66]
plt.hist(x,label="first plot",linewidth=4)
plt.hist(x2,label="second plot",linewidth=6)
plt.hist(y,label="second plot",linewidth=6)
plt.hist(y2,label="second plot",linewidth=6)
plt.legend()  # it is important to call this method legend() to get the labels displayed
plt.show()
```
